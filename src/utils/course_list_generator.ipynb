```python
index = [f"{i+1:02d}" for i in range(len(course_names))]



courses_pd = pd.DataFrame(
	{'Course Code': list(course_names.keys()), 'Course Name': list(course_names.values())},
	index=index
)

# add semester and year columns by mapping with ee_courses and cs_ai_courses
def add_semester_year(courses_pd, course_map):
    if 'Semester' not in courses_pd.columns:
        courses_pd['Semester'] = None
    if 'Year' not in courses_pd.columns:
        courses_pd['Year'] = None


    for idx, code in zip(courses_pd.index, courses_pd['Course Code']):
        # Only update if values are not already set
        if pd.isna(courses_pd.at[idx, 'Semester']) and pd.isna(courses_pd.at[idx, 'Year']):
            for y, semesters in course_map.items():
                for s, codes in semesters.items():
                    if code in codes:
                        courses_pd.at[idx, 'Semester'] = s
                        courses_pd.at[idx, 'Year'] = y
                        break
                else:
                    continue
                break

 
add_semester_year(courses_pd, ee_courses)
add_semester_year(courses_pd, cs_ai_courses)

```

```python
def set_required(courses_pd, required_courses, column_name):
	courses_pd[column_name] = courses_pd['Course Code'].apply(lambda x: 1 if x in required_courses else 0)
 
set_required(courses_pd, cs_ai_required, "CS/AI Required")
set_required(courses_pd, ee_required, "EE Required")
```
